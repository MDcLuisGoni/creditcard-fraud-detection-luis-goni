```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-whitegrid")
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")
```

```python
DATA_PATH = "creditcard.csv"  
df = pd.read_csv("../creditcard.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nInfo:")
df.info()
```

```python
df.describe().T
```

```python
class_counts = df["Class"].value_counts().sort_index()
class_counts
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.xticks([0, 1], ["No Fraude", "Fraude"])
plt.ylabel("Cantidad de transacciones")
plt.title("DistribuciÃ³n de la variable Class")
plt.show()

fraud_rate = class_counts[1] / class_counts.sum()
print(f"Tasa de fraude: {fraud_rate:.4%}")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Amount"], bins=100, kde=False)
plt.title("DistribuciÃ³n del monto (Amount)")
plt.xlabel("Amount")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Amount"])
plt.title("Boxplot de Amount")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Time"], bins=100, kde=False)
plt.title("DistribuciÃ³n de Time")
plt.xlabel("Time (segundos desde la primera transacciÃ³n)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0, square=True, cbar_kws={"shrink": 0.5})
plt.title("Matriz de correlaciÃ³n - Todas las variables")
plt.show()
```
